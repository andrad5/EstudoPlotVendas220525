# Explorando a Visualização de Dados com Python:
* **Uma Jornada Prática e um Estudo de Caso de Vendas**

Um espaço onde explorei o poder da **visualização de dados em Python**. Meu objetivo inicial foi aprender e dominar diversas técnicas de plotagem utilizando as bibliotecas `Matplotlib` e `Seaborn`.

Ao longo deste notebook, você verá exemplos práticos dos seguintes tipos de gráficos, cada um com suas particularidades e aplicações:

**Gráficos de Linha:** Ideais para mostrar tendências e mudanças ao longo do tempo.

**Gráficos de Dispersão (Scatter Plots):**
* Perfeitos para visualizar a relação entre duas variáveis numéricas.

**Histogramas:**
* Excelentes para entender a distribuição e a frequência de um conjunto de dados.

**Gráficos de Barras (Verticais e Horizontais):**
* Ótimos para comparar quantidades entre diferentes categorias.

**Gráficos de Pizza:**
 *Úteis para mostrar a proporção de partes em relação a um todo.

**Mapas de Calor (Heatmaps):**
* Poderosos para visualizar correlações e padrões em matrizes de dados.

Após essa exploração prática de cada tipo de gráfico, acabei unindo tudo em um estudo de caso real simulado, onde esse estudo final demonstra como aplicar essas visualizações para extrair insights de negócio concretos e tomar decisões informadas.

# 1. Grafico de linha


```python
import matplotlib.pyplot as plt

# 1 - Criamos as variaveis
x = [1,2,3,4,5]
y = [10,12,5,8,10]

# 2 - Geramos os graficos
plt.plot(x,y)

# 2.1 - Titulo do gráfico
plt.title('Meu titulo¹')

# 2.2 - dsc eixo x
plt.xlabel('Titulo eixo x²')

# 2.3 - dsc eixo y
plt.ylabel('Titulo eixo y³')

# 3
plt.show()
```

# 2. Grafico de barras

```python
import matplotlib.pyplot as plt

categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 40, 80]

plt.bar(categorias, valores, width=0.9)
plt.title('categorias X valores')
plt.xlabel('Meu titulo eixo X')
plt.ylabel('Meu titulo eixo Y')
plt.show()
```

# 2.1. Grafico de barras horizontais

```python
import matplotlib.pyplot as plt
import numpy as np

produtos = ['Arroz', 'Feijão', 'Macarrão', 'Óleo', 'Leite']
valor = [150, 120, 80, 60, 90]

#plt.bar(produtos, valor)
plt.barh(produtos, valor,
         color='skyblue',
         edgecolor='darkblue',
         height=0.9)

plt.title('Vendas por Produto',
          fontsize=14,
          fontweight='bold')

plt.ylabel('Produto',
           fontsize=12)

plt.xlabel('Vendas',
           fontsize=12)
```

# 3. Histograma(personalizado)

```python
import matplotlib.pyplot as plt
import numpy as np

df_hist = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.hist(df_hist,
         bins=100,  #Numero de barras que add
         color='yellowgreen', #Cor da barra
         alpha=0.5, #Transparencia
         edgecolor='darkseagreen', #Cor da borda
         width=0.1) #Largura das barras

plt.title('Meu histograma')
plt.xlabel('Valores')
plt.ylabel('Frequencia')
plt.grid(True)
plt.show()
```

A partir daqui, começamos a ultilizar algumas personalizações, como alterar cor, tamanho e ate borda, segue oque foi usado nesse grafico:

1. **np.random.randn(n)**: Usado para criar um dataframe com 1000 numeros aleatorios, mas no seu caso, pode ser seu proprio df. Esses numeros normalmente são ultilizados em simulações, teste de hipoteses ou para gerar dados de exemplo.

2. **plt.figure(figsize=(x,y))**: Ele altera o tamanho do grafico, podemos ver que ele esta deixando 10 de largura por 6 de altura.

3. **bins**: Ele define a quantidade de barras que quero no meu histograma, nesse ele esta definido para 100 barras.

4. **color**: Ultilizado para alterar a cor do grafico, aqui esta configurado para 'yellowgreen".
Algumas cores para usar de exemplo:

colors plt.webp

5. **alpha**: Ultilizado para transparencia do gráfico, onde os valores vão de 0.0, totalmente opaco, e 1.0 totalmente transparente. Nesse caso esta configurado para 50% de transparencia(0.5).

6. **edgecolor**: Ultilizado para definir as cores das bordas das colunas, nesse caso, esta configurado para darkseagreen

7. **width**: Ultilizado para deternumar a largra das barras verticais, caso queira definir a largura das barras horizontais, ultilizamos 'height'.

Nesse caso usei apenas os mais comuns, mas pode haver muito mais personalizações na [documentação](https://matplotlib.org/stable/).

#  4. Gráfico de dispersão

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(50)
y = 2 * x + 1 + 0.1 * np.random.randn(50)


plt.scatter(x,y, color='purple')
plt.title('Grafico de dispersão')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()

```

# 5. Gráfico de pizza

```python
import matplotlib.pyplot as plt

categorias = 'Categoria A', 'Categoria B', 'Categoria C'
vlr = [150, 150, 100]

plt.pie(vlr, labels=categorias, autopct='%0.2f%%')
plt.title('Gráfico de pizza')
plt.show()
#startangle=90
```

# 6. Mapa de Calor
**Esse pode ser um pouco mais complexo, vamos para alguns detalhes:**

 * annot=True: Usado para exibir os valores nas celulas
 * cmap='coolwarm': Usado para definir a paleta de cores, azul para negativo, vermelho para positivo
 * fmt=".2f": Usado para formatar os números exibidos com duas casas decimais
 * linewidths=.5: Usado para adicionar linhas entre as células
 * plt.xticks(rotation=45, ha='right'): Usado para rotação dos rótulos do eixo X
 * plt.yticks(rotation=0): Usado para rotação dos rótulos do eixo Y
 * plt.tight_layout(): Usado para ajusta o layout para que os rótulos não sejam cortados

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = np.random.rand(10,10)

plt.figure(figsize=(8, 7))
sns.heatmap(df,
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            linewidths=.5)
plt.title('Matriz de Correlação das Métricas de Marketing', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

7. Mapa de area

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y1 = [10,20,30,40,50]
y2 = [25,30,45,50,60]

plt.fill_between(x, y1,y2, alpha=0.4)
plt.title('Mapa de area')
plt.show()
```

# **Análise de Vendas Online: Desvendando o Desempenho do 1º Trimestre de 2025**

* Este notebook apresenta uma análise detalhada das vendas online de nossa loja de roupas nos primeiros três meses de 2025 (Janeiro, Fevereiro e Março). O objetivo principal é identificar padrões, tendências e áreas de oportunidade para otimização da performance do negócio.

* Utilizaremos as bibliotecas `Matplotlib` e `Seaborn` em Python para criar visualizações claras e extrair insights acionáveis sobre a tendência de vendas mensais, o desempenho por categoria de produto, a satisfação do cliente e as correlações entre métricas de marketing.

----
# Criação de dataframe
A seguir vamos:
 1. Criar um df para aplicar os graficos na pratica
 2. Criar graficos a partir desse df


```python
import pandas as pd
import numpy as np

vendas_mensais = {'Mes': ['Janeiro', 'Fevereiro', 'Março'],
                'Total_Vendas': [125000, 118000,142000]
}
df_vendas_mensais = pd.DataFrame(vendas_mensais)
#df_vendas_mensais.head()

vendas_por_categoria = {
    'Categoria': ['Camisetas', 'Calças', 'Vestidos', 'Acessórios', 'Calçados'],
    'Receita_BRL': [55000, 48000, 35000, 20000, 30000]
}
df_vendas_categoria = pd.DataFrame(vendas_por_categoria)
#df_vendas_categoria.head()


np.random.seed(42)
satisfacao_clientes = np.random.normal(loc=4.0, scale=0.8, size=1000)
satisfacao_clientes = np.clip(satisfacao_clientes, 1, 5)

matriz_correlacao = np.array([
    [1.0, 0.7, 0.6, 0.4],
    [0.7, 1.0, 0.5, 0.8],
    [0.6, 0.5, 1.0, 0.3],
    [0.4, 0.8, 0.3, 1.0]
])

nomes_variaveis = ['Vendas', 'Cliques Anúncio', 'Conversão', 'Gasto Marketing']
df_correlacao = pd.DataFrame(matriz_correlacao, index=nomes_variaveis, columns=nomes_variaveis)
#df_correlacao.head()
```

------------------------------------------------
# **Gráfico de Linha:** Tendência de Vendas Mensais (Janeiro-Março 2025)
**Esse gráfico é sua bússola da performance ao longo do tempo. Ele mostra como o faturamento total da loja se comportou mês a mês.**

# Insights Chave:

**Queda em Fevereiro:** Houve uma leve queda nas vendas de Janeiro para Fevereiro, isso pode ser sazonal (fim das férias, volta às aulas, etc.), mas merece investigação.


* Perguntas a fazer:
 * Houve alguma campanha de marketing específica em janeiro que não se repetiu em fevereiro?
 * Algum feriado ou evento que impactou o consumo?
 * A loja lançou menos produtos novos em fevereiro?


**Recuperação e Crescimento em Março:** As vendas em Março apresentaram uma forte recuperação e o maior volume do trimestre. Isso é positivo e indica que a tendência final do trimestre é de crescimento.

**Oportunidade:** O que foi feito em Março que pode ser replicado? Campanhas de Páscoa, promoções específicas, ou um aumento no investimento em marketing podem ter contribuído.

**Visão Trimestral:** A loja faturou um total de R$ 385.000 no primeiro trimestre. Março representa um bom impulso para o próximo trimestre.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.figure(figsize=(9,5))
plt.plot(df_vendas_mensais['Mes'],
         df_vendas_mensais['Total_Vendas'],
         marker='o',
         linestyle='-',
         color='purple',
         linewidth=2)
plt.title('Tendência de Vendas Mensais (Jan-Mar 2025)')
plt.xlabel('Mês')
plt.ylabel('Total de Vendas')
plt.grid(True)
plt.show()
```

----
# **Gráfico de Barras Horizontal:** Receita por Categoria de Produto (Trimestre)

**Este gráfico é o seu mapa dos produtos mais lucrativos. Ele compara diretamente o desempenho de cada categoria.**

# Insights Chave:

**Campeões de Vendas:** As Camisetas e Calças são as categorias que mais geraram receita. São os "carro-chefe" da loja.


**Ação:** Garanta que essas categorias estejam sempre bem estocadas e com boa visibilidade no site. Considere expandir a variedade de modelos ou lançar coleções especiais.

**Categorias com Potencial:** Vestidos e Calçados mostram uma receita considerável, mas com margem para crescimento.


**Ação:** Avalie a taxa de conversão dessas categorias, o tráfego que recebem e se há oportunidades para campanhas de marketing direcionadas ou promoções.

Área para Otimização: Acessórios é a categoria com menor receita.

**Perguntas a fazer:**

 * Os acessórios são vistos como um complemento ou um produto principal?
 * A margem de lucro é alta o suficiente para justificar a receita?
 * Há pouca variedade?
 * Onde eles são exibidos no site?


```python

plt.figure(figsize=(10, 6))
plt.barh(df_vendas_categoria['Categoria'],
         df_vendas_categoria['Receita_BRL'],
         color='lightgreen',
         edgecolor='darkgreen',
         height=0.7)

plt.xlabel('Receita (BRL)', fontsize=12)
plt.ylabel('Categoria de Produto', fontsize=12)
plt.title('Receita por Categoria de Produto (Trimestre)', fontsize=15, fontweight='bold', color='darkgreen')

plt.gca().invert_yaxis()

for i, v in enumerate(df_vendas_categoria['Receita_BRL']):
    plt.text(v + 1000, i, f'R$ {v:,.0f}', color='black', va='center', fontsize=10) # Formata com virgula
plt.grid(axis='x', linestyle=':', alpha=0.6)
plt.show()
```

----
# **Histograma:** Distribuição da Satisfação do Cliente

**Este histograma mostra a opinião geral dos seus clientes e a frequência de cada nível de satisfação.**

# Insights Chave:

**Alta Satisfação Média:** A maioria das avaliações está concentrada entre 3.5 e 4.5 estrelas, com um pico claro na faixa de 4.0 a 4.5 estrelas. Isso indica que a maioria dos clientes está satisfeita ou muito satisfeita com a loja. Isso é um ponto forte!


**Oportunidade:** Colete depoimentos e use as avaliações positivas como prova social em campanhas de marketing. Incentive mais clientes satisfeitos a deixar avaliações.


**Poucas Avaliações Baixas:** Há uma baixa frequência de avaliações de 1 a 2.5 estrelas, o que é um bom sinal de que problemas graves são raros.


**Ação:** Monitore de perto as poucas avaliações negativas para entender as causas raiz e resolver esses problemas pontuais rapidamente, transformando uma experiência ruim em uma oportunidade de fidelização.


**Espaço para Excelência:** Embora a satisfação seja alta, ainda há uma pequena cauda de avaliações entre 2.5 e 3.5 estrelas.


**Ação:** Analise os comentários dessas avaliações "neutras" para identificar áreas de melhoria contínua, como a experiência de entrega, qualidade de produtos específicos ou suporte ao cliente.

**O objetivo é mover esses clientes para a faixa de 4-5 estrelas.**

```python
plt.figure(figsize=(9, 6))
plt.hist(satisfacao_clientes, bins=np.arange(1, 5.5, 0.5), color='gold', edgecolor='orange', alpha=0.8)
plt.title('Distribuição da Satisfação do Cliente (1-5 Estrelas)', fontsize=16, fontweight='bold')
plt.xlabel('Avaliação (Estrelas)', fontsize=12)
plt.ylabel('Número de Clientes', fontsize=12)
plt.xticks(np.arange(1, 5.5, 0.5)) # Garante que os ticks estejam nos limites dos bins
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# **Mapa de Calor:** Correlação entre Métricas de Marketing
**Este mapa de calor é uma ferramenta de diagnóstico para entender como suas ações de marketing e resultados se relacionam.**

# Insights Chave:

* **Forte Correlação com Cliques no Anúncio:**

* **Vendas e Cliques no Anúncio (0.7):** Há uma correlação forte e positiva entre o número de cliques nos anúncios e as vendas. Isso é intuitivo: mais cliques significam mais tráfego, o que geralmente leva a mais vendas.
** **Ação:** Continue otimizando suas campanhas para gerar mais cliques de qualidade. Invista em copy (texto) e criativos (imagens/vídeos) que convertam bem.

* **Gasto com Marketing e Cliques no Anúncio (0.8):** O gasto com marketing tem uma correlação muito forte e positiva com os cliques. Isso sugere que o investimento em marketing está efetivamente trazendo tráfego para o seu site.
** Ação: Se o retorno sobre o investimento (ROI) for positivo, considere aumentar o orçamento de marketing, pois parece estar diretamente ligado ao aumento de cliques.

* **Correlação Moderada/Fraca com Conversão:**
* **Vendas e Conversão (0.6):** A correlação é moderada. Embora a conversão seja importante para vendas, outros fatores também estão influenciando.

* **Cliques no Anúncio e Conversão (0.5):** A correlação é um pouco mais fraca. Isso pode indicar que, embora você esteja atraindo cliques, a qualidade do tráfego ou a experiência no site podem não estar otimizadas para converter esses cliques em vendas.

 * **Ação:** Foco em otimização da taxa de conversão (CRO). Verifique o design do site, a clareza das descrições dos produtos, o processo de checkout, a velocidade da página. Talvez os anúncios estejam atraindo pessoas que não são o público-alvo ideal.

* **Gasto com Marketing:**

* Sua principal correlação é com os cliques (0.8), o que é bom. Mas a correlação com as vendas (0.4) e, especialmente, com a conversão (0.3) é mais fraca.

 * **Ação:** Reavalie a qualidade do gasto de marketing. É possível que o investimento esteja gerando muitos cliques, mas nem todos se traduzem em clientes pagantes. Foco em segmentação de público e canais que trazem clientes mais qualificados.

```python
plt.figure(figsize=(8, 7))
sns.heatmap(df_correlacao,
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            linewidths=.5)
plt.title('Matriz de Correlação das Métricas de Marketing', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
